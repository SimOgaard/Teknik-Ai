## Generate Average

```python
def Generate_Average(Inputdata, Parameter,setsize):
    subsets=[]
    Newcolumn=[]
    Outputdata= pd.DataFrame()
    rest= setsize% 2 #Mudelo
    i = int(setsize/2)+rest
    max= len(Inputdata)
    while i < max:
        subset=Inputdata[i-int(setsize/2)   :   i+int(setsize/2)]
        average = subset[Parameter].mean()
        Newcolumn.append(average)
        i += 1
    Outputdata['Average'] = Newcolumn
    Outputdata['time'] = Inputdata.time
    Outputdata['index_col'] = Outputdata.index
    print(len(Outputdata))
    return Outputdata

def Generate_Average2(Inputdata, InputParameters,OutputParameters,setsize):
    Outputdata= pd.DataFrame()
    Paracounter=0
    for Parameter in InputParameters:
        
        Newcolumn=[]
        rest= setsize% 2 #Mudelo
        i = int(setsize/2)+rest
        max= len(Inputdata)
        while i < max:
            subset=Inputdata[i-int(setsize/2):i+int(setsize/2)]
            average = subset[Parameter].mean()
            Newcolumn.append(average)
            i += 1
            
        Outputdata[OutputParameters[Paracounter]] = Newcolumn
        Paracounter+=1
    Outputdata['time'] = Inputdata.time
    Outputdata['index_col'] = Outputdata.index
    print(len(Outputdata))
    return Outputdata


# Averageofset= Generate_Average(rawdata,"acc-side", 14)
# Averageofset.head(15)

Averagedata= Generate_Average2(rawdata,['acc-side','acc-up/down','acc-front/back'],['Average-side','Average-up/down','Average-front/back'], 9)
Averagedata.head()

plotter(Averagedata,['Average-side','Average-up/down','Average-front/back']);

df= Generate_Average2(Averagedata,['Average-side','Average-up/down','Average-front/back'],['AAS','AAUD','AAFB'], 12)
df.head()

# plotter(df,['AverageofAverage-side','AverageofAverage-up/down','AverageofAverage-front/back']);
```

## Count local min and max

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate a noisy AR(1) sample
np.random.seed(0)
rs = np.random.randn(200)
xs = [0]
for r in rs:
    xs.append(xs[-1]*0.9 + r)
df = pd.DataFrame(xs, columns=['data'])


# Find local peaks
df['min'] = df.data[(df.data.shift(1) > df.data) & (df.data.shift(-1) > df.data)]
df['max'] = df.data[(df.data.shift(1) < df.data) & (df.data.shift(-1) < df.data)]

# Plot results
plt.scatter(df.index, df['min'], c='r')
plt.scatter(df.index, df['max'], c='g')
df.data.plot()

#Hur kan man ändra för att göra likadant med er data?
```
