```python
import pandas as pd
rawdata = pd.read_csv( "rawdata119870.csv")
rawdata['index_col'] = rawdata.index
Smallpartofdata=rawdata[150:600]

def plotter(data, parameters):
    for parameter in parameters:
        data[parameter].plot()

def Generate_Average(Inputdata, InputParameters,OutputParameters, setsize):
    Outputdata=pd.DataFrame()
    Paracounter=0
    for Parameter in InputParameters:
        Newcolumn=[]
        rest= setsize% 2
        i = int(setsize/2)+rest
        max= len(Inputdata)
        while i < max:
            subset=Inputdata[i-int(setsize/2):i+int(setsize/2)]
            average = subset[Parameter].mean()
            Newcolumn.append(average)
            i += 1
        Outputdata[OutputParameters[Paracounter]] = Newcolumn
        Paracounter+=1
    Outputdata['time'] = Inputdata.time
    Outputdata['index_col'] = Outputdata.index
    return Outputdata

Averagedata= Generate_Average(Smallpartofdata,['acc-side','acc-up/down','acc-front/back'],['Average-side','Average-up/down','Average-front/back'], 9)
AverageofAveragedata= Generate_Average(Averagedata,['Average-side','Average-up/down','Average-front/back'],['AverageofAverage-side','AverageofAverage-up/down','AverageofAverage-front/back'], 9)
plotter(AverageofAveragedata,['AverageofAverage-side','AverageofAverage-up/down','AverageofAverage-front/back']);

df = AverageofAveragedata
data = 'acc-up/down'
df['min'] = df.data[(df.data.shift(1) > df.data) & (df.data.shift(-1) > df.data)]
df['max'] = df.data[(df.data.shift(1) < df.data) & (df.data.shift(-1) < df.data)]

plt.scatter(df.index, df['min'], c='r')
plt.scatter(df.index, df['max'], c='g')
df.data.plot()
```
